```python
x=dataset[['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'x', 'y','z' ]]
x
```

```python
y=dataset[['price']]
y
```

```python
x=pd.get_dummies(x)
```

```python
x.tail()
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(xtrain,ytrain)
```

```python
y_pred=lr_model.predict(xtest)
```

```python
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),ytest.values.reshape(len(ytest),1)),1))
```

```python
from sklearn.metrics import r2_score
print("The score of Linear Regression is {:.2f}%".format(r2_score(y_pred,ytest)*100))
```

```python
from sklearn.ensemble import RandomForestRegressor
rm_model=RandomForestRegressor(n_estimators=10)
rm_model.fit(xtrain,ytrain)
```

```python
y_pred2=rm_model.predict(xtest)
```

```python
np.set_printoptions(precision=2)
print(np.concatenate((y_pred2.reshape(len(y_pred2),1),ytest.values.reshape(len(ytest),1)),1))
```

```python
print("Score of Random forest model is {:.2f}%".format(r2_score(y_pred2,ytest)*100))
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model=DecisionTreeRegressor()
dt_model.fit(xtrain,ytrain)
```

```python
y_pred3=dt_model.predict(xtest)
```

```python
np.set_printoptions(precision=2)
print(np.concatenate((y_pred3.reshape(len(y_pred3),1),ytest.values.reshape(len(ytest),1)),1))
```
